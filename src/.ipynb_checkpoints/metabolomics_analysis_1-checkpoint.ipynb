```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import pandas as pd
from sklearn.preprocessing import RobustScaler, StandardScaler
from sklearn.decomposition import PCA
from sklearn.pipeline import make_pipeline
from sklearn.manifold import TSNE
from statsmodels.stats.multitest import multipletests
import scipy.stats as stats
```

# Read in Raw Data

```python
df_metabolite = pd.read_csv("../raw/metabolon_data.csv")
df_metabolite.drop(columns=["Sample"],inplace=True)
df_metabolite.head()
```

```python
# split into features and sample IDs
df_metab_features = df_metabolite.drop(columns=["Group"])
df_metab_group = df_metabolite.Group
df_metab_features.head()
```

```python
# print out the unique experimental groups
print(np.unique(df_metab_group))

df_metab_group.head()
```

```python
# create dictionary to convert chemical IDs in header to metabolite name/info
# make chemical ID as string, so both BIOCHEMICAL and CHEMICAL ID are strings
df_chemIDs = pd.read_csv("../raw/metabolite_definitions.csv")
df_chemIDs.head()
```

```python
dict_chemID_to_metabolite = pd.Series(df_chemIDs["BIOCHEMICAL"].values, 
                                      index=df_chemIDs['CHEMICAL ID'].apply(str)).to_dict()
dict_chemID_to_compID = pd.Series(df_chemIDs["COMP ID"].values, 
                                      index=df_chemIDs['CHEMICAL ID'].apply(str)).to_dict()

# test dictionary
print(dict_chemID_to_metabolite["100002945"])
print(dict_chemID_to_compID["100002945"])
```

```python
# construct chemID to pubchem ID
df_pubchem = pd.read_csv('../raw/compIDtoPubchemID.csv',dtype={'compID':str,
                                                               'pubchemID':str})
df_pubchem.head()
dict_chemID_to_pubchem = pd.Series(df_pubchem["pubchemID"].values, 
                                      index=df_pubchem['compID']).to_dict()
```

```python
def chemIDtoPubchem(dict_chemToComp, dict_compToPubchem, key):
    return dict_compToPubchem[str(dict_chemToComp[key])]

# test dictionary fcn to go from chemID to pubchem ID
print(chemIDtoPubchem(dict_chemID_to_compID,
                      dict_chemID_to_pubchem,
                      "100002945"))
```

# Begin Scaling and Dimensionality Reduction

## Begin for all data points

```python
# Make pipelines
pipeline_1 = make_pipeline(RobustScaler(), PCA())
pipeline_2 = make_pipeline(RobustScaler(), TSNE(n_components=3, random_state=42))

# PCA
df_metab_features_pca = pipeline_1.fit(df_metab_features)
print(df_metab_features_pca[1].components_.shape) # num_components, num_features

# tSNE
df_metab_features_tSNE = pipeline_2.fit(df_metab_features)
print(df_metab_features_tSNE[1].embedding_.shape) # num_samples, num_components

# transform dim reduction onto feature matrix
# output is num_samples x num_components
df_features_pca_transf = pd.DataFrame(df_metab_features_pca[1].transform(df_metab_features))
print(df_features_pca_transf.shape)

df_features_tSNE_transf = pd.DataFrame(df_metab_features_tSNE[1].fit_transform(df_metab_features))
print(df_features_tSNE_transf.shape)
```

```python
def plot_DimRed(df_pca_transf, df_tSNE_transf, df_groupNames,
                pca_pipeline, tSNE_pipeline,
                fn_out_explVar, fn_out_PCA2D, fn_out_PCA3D,
               fn_out_tSNE2D, fn_out_tSNE3D):
    # plot explained variance
    plt.figure()
    expl_var_array = np.cumsum(pca_pipeline[1].explained_variance_ratio_)
    plt.plot(expl_var_array)
    plt.ylabel('Fraction of Variance Explained')
    plt.xlabel('# of Components')
    plt.savefig(fn_out_explVar)
    plt.show()


    ###########################################################################################
    ## plot PCA
    # define colormap
    n = len(np.unique(df_groupNames))
    color=iter(plt.cm.rainbow(np.linspace(0,1,n)))

    # plot 2D
    plt.figure()
    for i in range(n):
        c = next(color)
        group = np.unique(df_groupNames)[i]
        idx = np.where(df_groupNames == group)[0]
        plt.scatter(df_pca_transf.loc[idx,0],
                    df_pca_transf.iloc[idx,1],
                   c=c, label=group, alpha = 0.7)

    plt.ylabel('PC2: {}'.format(round(expl_var_array[1]-expl_var_array[0],3)))
    plt.xlabel('PC1: {}'.format(round(expl_var_array[0],3)))
    plt.xticks([])
    plt.yticks([])
    plt.legend()
    plt.title('PCA of all Timepoints')
    plt.savefig(fn_out_PCA2D)
    plt.show()

    # plot 3D PCA
    # define colormap
    n = len(np.unique(df_metab_group))
    color=iter(plt.cm.rainbow(np.linspace(0,1,n)))

    fig = plt.figure()
    ax = Axes3D(fig)
    for i in range(n):
        c = next(color)
        group = np.unique(df_groupNames)[i]
        idx = np.where(df_groupNames == group)[0]
        ax.scatter(df_pca_transf.loc[idx,0],
                    df_pca_transf.iloc[idx,1],
                   df_pca_transf.iloc[idx,2],
                   c=c, label=group)

    ax.set_xlabel('PC1: {}'.format(round(expl_var_array[0],3)))
    ax.set_ylabel('PC2: {}'.format(round(expl_var_array[1]-expl_var_array[0],3)))
    ax.set_zlabel('PC3: {}'.format(round(expl_var_array[2]-expl_var_array[1]-expl_var_array[0],3)))
    ax.set_xticklabels([])
    ax.set_yticklabels([])
    ax.set_zticklabels([])
    plt.legend()
    plt.title('PCA of all Timepoints')
    plt.savefig(fn_out_PCA3D)
    plt.show()

    ###########################################################################################
    ## plot tSNE
    # define colormap
    n = len(np.unique(df_groupNames))
    color=iter(plt.cm.rainbow(np.linspace(0,1,n)))

    # plot 2D
    plt.figure()
    for i in range(n):
        c = next(color)
        group = np.unique(df_groupNames)[i]
        idx = np.where(df_groupNames == group)[0]
        plt.scatter(df_tSNE_transf.iloc[idx,0],
                    df_tSNE_transf.iloc[idx,1],
                   c=c, label=group, alpha = 0.7)

    plt.xlabel('tSNE1')
    plt.ylabel('tSNE2')
    plt.xticks([])
    plt.yticks([])
    plt.legend()
    plt.title('tSNE of all Timepoints')
    plt.savefig(fn_out_tSNE2D)
    plt.show()

    # plot 3D PCA
    # define colormap
    n = len(np.unique(df_groupNames))
    color=iter(plt.cm.rainbow(np.linspace(0,1,n)))

    fig = plt.figure()
    ax = Axes3D(fig)
    for i in range(n):
        c = next(color)
        group = np.unique(df_groupNames)[i]
        idx = np.where(df_groupNames == group)[0]
        ax.scatter(df_tSNE_transf.iloc[idx,0],
                    df_tSNE_transf.iloc[idx,1],
                   df_tSNE_transf.iloc[idx,2],
                   c=c, label=group)

    ax.set_xlabel('tSNE1')
    ax.set_ylabel('tSNE2')
    ax.set_zlabel('tSNE3')
    ax.set_xticklabels([])
    ax.set_yticklabels([])
    ax.set_zticklabels([])
    plt.legend()
    plt.title('tSNE of all Timepoints')
    plt.savefig(fn_out_tSNE3D)
    plt.show()
```

## Perform dim reduction with one factor held constant

We have three factors: 1) Time (Early or Late), 2) Genotype (WT or KO), and 3) Disease state (Healthy or Sick). Repeat analysis holding one of these constant

```python
def split_dfs(df_in, list_1, list_2):
    df_1 = df_in.loc[df_in['Group'].isin(list_1)]
    df_1_features = df_1.drop(columns=["Group"])
    df_1_group = df_1.Group
    
    df_2 = df_in.loc[df_in['Group'].isin(list_2)]
    df_2_features = df_2.drop(columns=["Group"])
    df_2_group = df_2.Group
    
    return list([df_1_features, df_1_group, df_2_features, df_2_group])

def run_pipelines(df_features, pipeline_1, pipeline_2):
    # PCA
    df1 = pipeline_1.fit(df_features)
    print(df1[1].components_.shape) # num_components, num_features

    # tSNE
    df2 = pipeline_2.fit(df_features)
    print(df2[1].embedding_.shape) # num_samples, num_components
    
    return df1, df2
    
```

```python
################################################################################
### Time
print("Time")
# split into features and sample IDs
dfList_Time = split_dfs(df_metabolite, list(["EH_WT","ES_KO","ES_WT"]),
                                           list(["LH_WT","LS_KO","LS_WT"]))

df_Time_1_PCA, df_Time_1_tSNE = run_pipelines(dfList_Time[0], pipeline_1, pipeline_2)
df_Time_2_PCA, df_Time_2_tSNE = run_pipelines(dfList_Time[2], pipeline_1, pipeline_2)

### transform dim reduction onto feature matrix
# output is num_samples x num_components
df_Time_1_pca_transf = pd.DataFrame(df_Time_1_PCA[1].transform(dfList_Time[0]))
print(df_Time_1_pca_transf.shape)

df_Time_1_tSNE_transf = pd.DataFrame(df_Time_1_tSNE[1].fit_transform(dfList_Time[0]))
print(df_Time_1_tSNE_transf.shape)

df_Time_2_pca_transf = pd.DataFrame(df_Time_1_PCA[1].transform(dfList_Time[2]))
print(df_Time_2_pca_transf.shape)

df_Time_2_tSNE_transf = pd.DataFrame(df_Time_2_tSNE[1].fit_transform(dfList_Time[2]))
print(df_Time_2_tSNE_transf.shape)



################################################################################
### Genotype
print("Genotype")
dfList_Geno = split_dfs(df_metabolite, list(["EH_WT","ES_WT","LH_WT","LS_WT"]),
                                           list(["ES_KO","LS_KO"]))

df_Geno_1_PCA, df_Geno_1_tSNE = run_pipelines(dfList_Geno[0], pipeline_1, pipeline_2)
df_Geno_2_PCA, df_Geno_2_tSNE = run_pipelines(dfList_Geno[2], pipeline_1, pipeline_2)

### transform dim reduction onto feature matrix
# output is num_samples x num_components
df_Geno_1_pca_transf = pd.DataFrame(df_Geno_1_PCA[1].transform(dfList_Geno[0]))
print(df_Geno_1_pca_transf.shape)

df_Geno_1_tSNE_transf = pd.DataFrame(df_Geno_1_tSNE[1].fit_transform(dfList_Geno[0]))
print(df_Geno_1_tSNE_transf.shape)

df_Geno_2_pca_transf = pd.DataFrame(df_Geno_1_PCA[1].transform(dfList_Geno[2]))
print(df_Geno_2_pca_transf.shape)

df_Geno_2_tSNE_transf = pd.DataFrame(df_Geno_2_tSNE[1].fit_transform(dfList_Geno[2]))
print(df_Geno_2_tSNE_transf.shape)

################################################################################
### Disease State
print("Disease")
dfList_Disease = split_dfs(df_metabolite, list(["EH_WT","LH_WT"]),
                                           list(["ES_WT", "ES_KO", "LS_WT","LS_KO"]))

df_Disease_1_PCA, df_Disease_1_tSNE = run_pipelines(dfList_Disease[0], pipeline_1, pipeline_2)
df_Disease_2_PCA, df_Disease_2_tSNE = run_pipelines(dfList_Disease[2], pipeline_1, pipeline_2)

### transform dim reduction onto feature matrix
# output is num_samples x num_components
df_Disease_1_pca_transf = pd.DataFrame(df_Disease_1_PCA[1].transform(dfList_Disease[0]))
print(df_Disease_1_pca_transf.shape)

df_Disease_1_tSNE_transf = pd.DataFrame(df_Disease_1_tSNE[1].fit_transform(dfList_Disease[0]))
print(df_Disease_1_tSNE_transf.shape)

df_Disease_2_pca_transf = pd.DataFrame(df_Disease_2_PCA[1].transform(dfList_Disease[2]))
print(df_Disease_2_pca_transf.shape)

df_Disease_2_tSNE_transf = pd.DataFrame(df_Disease_2_tSNE[1].fit_transform(dfList_Disease[2]))
print(df_Disease_2_tSNE_transf.shape)
```

```python
def plot_DimRed(df_1_pca_transf, df_1_tSNE_transf, df_1_groupNames,
                df_2_pca_transf, df_2_tSNE_transf, df_2_groupNames,
                df_1_pca_pipeline, df_1_tSNE_pipeline,
                df_2_pca_pipeline, df_2_tSNE_pipeline,
                title_1, title_2, title_3,
                fn_out_explVar, fn_out_PCA2D, 
                fn_out_PCA3D_1, fn_out_PCA3D_2,
                fn_out_tSNE2D,
                fn_out_tSNE3D_1, fn_out_tSNE3D_2):
    
    # plot explained variance
    fig, axes = plt.subplots(nrows=1, ncols=2)
    expl_var_array_1 = np.cumsum(df_1_pca_pipeline[1].explained_variance_ratio_)
    expl_var_array_2 = np.cumsum(df_2_pca_pipeline[1].explained_variance_ratio_)
    axes[0].plot(expl_var_array_1)
    axes[0].set_xlabel('# of Components')
    axes[0].set_ylabel('Fraction of Variance Explained')
    axes[0].set_title(title_1)
    
    axes[1].plot(expl_var_array_2)
    axes[1].set_xlabel('# of Components')
    axes[1].set_ylabel('Fraction of Variance Explained')
    axes[1].set_title(title_2)
    plt.tight_layout()
    plt.savefig(fn_out_explVar)
    plt.show()


    ###########################################################################################
    ## plot PCA
    # define colormap
    n1 = len(np.unique(df_1_groupNames))
    n2 = len(np.unique(df_2_groupNames))
    color1=iter(plt.cm.rainbow(np.linspace(0,1,n1)))
    color2=iter(plt.cm.rainbow(np.linspace(0,1,n2)))
    
    # plot 2D
    fig, axes = plt.subplots(nrows=1, ncols=2)
    for i in range(n1):
        c = next(color1)
        group = np.unique(df_1_groupNames)[i]
        idx = np.where(df_1_groupNames == group)[0]
        axes[0].scatter(df_1_pca_transf.loc[idx,0],
                    df_1_pca_transf.iloc[idx,1],
                   c=c, label=group, alpha = 0.5)

    for i in range(n2):
        c = next(color2)
        group = np.unique(df_2_groupNames)[i]
        idx = np.where(df_2_groupNames == group)[0]
        axes[1].scatter(df_2_pca_transf.loc[idx,0],
                    df_2_pca_transf.iloc[idx,1],
                   c=c, label=group, alpha = 0.5)
    
    axes[0].set_xlabel('PC1: {}'.format(round(df_1_pca_pipeline[1].explained_variance_ratio_[0],3)))
    axes[0].set_ylabel('PC2: {}'.format(round(df_1_pca_pipeline[1].explained_variance_ratio_[1],3)))
    axes[1].set_xlabel('PC1: {}'.format(round(df_2_pca_pipeline[1].explained_variance_ratio_[0],3)))
    axes[1].set_ylabel('PC2: {}'.format(round(df_2_pca_pipeline[1].explained_variance_ratio_[1],3)))
    axes[0].set_xticks([])
    axes[0].set_yticks([])
    axes[1].set_xticks([])
    axes[1].set_yticks([])
    axes[0].legend()
    axes[1].legend()
    axes[0].set_title(title_1)
    axes[1].set_title(title_2)
    plt.suptitle(title_3)
    plt.savefig(fn_out_PCA2D)
    plt.show()

    # plot 3D PCA
    # define colormap
    n1 = len(np.unique(df_1_groupNames))
    n2 = len(np.unique(df_2_groupNames))
    color1=iter(plt.cm.rainbow(np.linspace(0,1,n1)))
    color2=iter(plt.cm.rainbow(np.linspace(0,1,n2)))

    fig = plt.figure()
    ax = Axes3D(fig)
    
    for i in range(n1):
        c = next(color1)
        group = np.unique(df_1_groupNames)[i]
        idx = np.where(df_1_groupNames == group)[0]
        ax.scatter(df_1_pca_transf.loc[idx,0],
                        df_1_pca_transf.iloc[idx,1],
                        df_1_pca_transf.iloc[idx,2],
                        c=c, label=group, alpha=1)

    ax.set_xlabel('PC1: {}'.format(round(df_1_pca_pipeline[1].explained_variance_ratio_[0],3)))
    ax.set_ylabel('PC2: {}'.format(round(df_1_pca_pipeline[1].explained_variance_ratio_[1],3)))
    ax.set_zlabel('PC3: {}'.format(round(df_1_pca_pipeline[1].explained_variance_ratio_[2],3)))
    ax.set_xticklabels([])
    ax.set_yticklabels([])
    ax.set_zticklabels([])
    plt.legend()
    plt.title(title_1)
    plt.savefig(fn_out_PCA3D_1)
    plt.show()
    
    fig = plt.figure()
    ax = Axes3D(fig)
    
    for i in range(n2):
        c = next(color2)
        group = np.unique(df_2_groupNames)[i]
        idx = np.where(df_2_groupNames == group)[0]
        ax.scatter(df_2_pca_transf.loc[idx,0],
                        df_2_pca_transf.iloc[idx,1],
                        df_2_pca_transf.iloc[idx,2],
                        c=c, label=group, alpha=1)

    ax.set_xlabel('PC1: {}'.format(round(df_2_pca_pipeline[1].explained_variance_ratio_[0],3)))
    ax.set_ylabel('PC2: {}'.format(round(df_2_pca_pipeline[1].explained_variance_ratio_[1],3)))
    ax.set_zlabel('PC3: {}'.format(round(df_2_pca_pipeline[1].explained_variance_ratio_[2],3)))
    ax.set_xticklabels([])
    ax.set_yticklabels([])
    ax.set_zticklabels([])
    plt.legend()
    plt.title(title_2)
    plt.savefig(fn_out_PCA3D_2)
    plt.show()

    ###########################################################################################
    ## plot tSNE
    # define colormap
    n1 = len(np.unique(df_1_groupNames))
    n2 = len(np.unique(df_2_groupNames))
    color1=iter(plt.cm.rainbow(np.linspace(0,1,n1)))
    color2=iter(plt.cm.rainbow(np.linspace(0,1,n2)))

    # plot 2D
    fig, axes = plt.subplots(nrows=1, ncols=2)
    for i in range(n1):
        c = next(color1)
        group = np.unique(df_1_groupNames)[i]
        idx = np.where(df_1_groupNames == group)[0]
        axes[0].scatter(df_1_tSNE_transf.loc[idx,0],
                    df_1_tSNE_transf.iloc[idx,1],
                   c=c, label=group, alpha = 0.5)

    for i in range(n2):
        c = next(color2)
        group = np.unique(df_2_groupNames)[i]
        idx = np.where(df_2_groupNames == group)[0]
        axes[1].scatter(df_2_tSNE_transf.loc[idx,0],
                    df_2_tSNE_transf.iloc[idx,1],
                   c=c, label=group, alpha = 0.5)

    axes[0].set_xlabel('tSNE1')
    axes[0].set_ylabel('tSNE2')
    axes[1].set_xlabel('tSNE1')
    axes[1].set_ylabel('tSNE2')
    axes[0].set_xticks([])
    axes[0].set_yticks([])
    axes[1].set_xticks([])
    axes[1].set_yticks([])
    axes[0].legend()
    axes[1].legend()
    axes[0].set_title(title_1)
    axes[1].set_title(title_2)
    plt.suptitle(title_3)
    plt.savefig(fn_out_tSNE2D)
    plt.show()

    # plot 3D tSNE
    # define colormap
    n1 = len(np.unique(df_1_groupNames))
    n2 = len(np.unique(df_2_groupNames))
    color1=iter(plt.cm.rainbow(np.linspace(0,1,n1)))
    color2=iter(plt.cm.rainbow(np.linspace(0,1,n2)))
    
    fig = plt.figure()
    ax = Axes3D(fig)
    
    for i in range(n1):
        c = next(color1)
        group = np.unique(df_1_groupNames)[i]
        idx = np.where(df_1_groupNames == group)[0]
        ax.scatter(df_1_tSNE_transf.loc[idx,0],
                        df_1_tSNE_transf.iloc[idx,1],
                        df_1_tSNE_transf.iloc[idx,2],
                        c=c, label=group, alpha=1)

    ax.set_xlabel('tSNE1')
    ax.set_ylabel('tSNE2')
    ax.set_zlabel('tSNE3')
    ax.set_xticklabels([])
    ax.set_yticklabels([])
    ax.set_zticklabels([])
    plt.legend()
    plt.title(title_1)
    plt.savefig(fn_out_tSNE3D_1)
    plt.show()
    
    fig = plt.figure()
    ax = Axes3D(fig)
    
    for i in range(n2):
        c = next(color2)
        group = np.unique(df_2_groupNames)[i]
        idx = np.where(df_2_groupNames == group)[0]
        ax.scatter(df_2_tSNE_transf.loc[idx,0],
                        df_2_tSNE_transf.iloc[idx,1],
                        df_2_tSNE_transf.iloc[idx,2],
                        c=c, label=group, alpha=1)

    ax.set_xlabel('tSNE1')
    ax.set_ylabel('tSNE2')
    ax.set_zlabel('tSNE3')
    ax.set_xticklabels([])
    ax.set_yticklabels([])
    ax.set_zticklabels([])
    plt.legend()
    plt.title(title_2)
    plt.savefig(fn_out_tSNE3D_2)
    plt.show()
```

```python
def plot_dfLoading(df_loadings, num_features, dictionary, title, fn_out):
    df_loadings = df_loadings.iloc[df_loadings['sum_2'].abs().argsort()[::-1]].reset_index(drop=True)
    n = list()
    for i in range(num_features):
        n.append(dictionary[df_loadings.Metab[i]])
    print(n)
    
    fig,ax = plt.subplots()
    ax.scatter(df_loadings.loc[:num_features, 'Comp1'],
               df_loadings.loc[:num_features, 'Comp2'])
    ax.set_xlabel('PC1')
    ax.set_ylabel('PC2')
    
    for i,txt in enumerate(n):
        print(df_loadings.loc[i, 'Comp1'])
        ax.annotate(txt, (df_loadings.loc[i, 'Comp1'],
                         df_loadings.loc[i, 'Comp2']))
    plt.title(title)
    plt.savefig(fn_out)
    plt.show()
```

# Perform Statistical Tests and volcano plots

```python
# Get combinations for 2-sample T-test
import itertools
```

```python
# log2 transform metabolite DataFrame
df_metab_features_log2 = np.log2(df_metab_features)

# Assemble list of comparisons
list_comparison = list(itertools.combinations(np.unique(df_metab_group), 2))
for comp in list_comparison:
    group1 = comp[0]
    group2 = comp[1]
    idx1 = df_metab_group.loc[df_metab_group==group1].index
    idx2 = df_metab_group.loc[df_metab_group==group2].index
    
    # compute log-fold-change of the different metabolites
    df_analysis1 = df_metab_features_log2.loc[idx1,:]
    df_analysis2 = df_metab_features_log2.loc[idx2,:]
    LFC = np.mean(df_analysis2, axis=0) / np.mean(df_analysis1, axis=0)

    # compute p-values of the metabolites
    pVals = stats.ttest_ind(df_analysis1, df_analysis2, axis=0)[1]
    
    # assemble dataFrame and write to csv
    df_stats = pd.DataFrame(data={'Metab':df_metab_features.columns,'LFC':LFC, 'pvalue':pVals})
    df_stats=df_stats.reset_index(drop=True)
    
    print(df_stats.shape)
    print(df_stats.head())
    
    df_stats.to_csv('../processed/stats_{}_to_{}.csv'.format(group2,group1))

    df_stats.dropna(axis=0,inplace=True) # remove rows with NA values
    print(df_stats.shape)

    # remove rows with LFC >10
    LFC_cutoff = 10
    idx_remove = df_stats[(df_stats['LFC']>LFC_cutoff) |
                          (df_stats['LFC']<-LFC_cutoff)].index
    print(idx_remove)
    print(df_stats.shape)
    df_stats.drop(index=idx_remove,inplace=True)
    print(df_stats.shape)
    
    # correct for multiple testing
    p_adj = multipletests(pvals=df_stats.pvalue, alpha=0.05, method="fdr_bh")
    cutoff_val = p_adj[3]
    df_stats['pvalue'] = p_adj[1]
    df_stats.to_csv('../processed/stats_{}_to_{}.csv'.format(group2,group1))

    # make CSV with significant results only, sorted in descending order by
    # absolute value of LFC
    df_stats_signif = df_stats.loc[df_stats['pvalue']<cutoff_val]
    df_stats_signif = df_stats_signif.iloc[df_stats_signif['LFC'].abs().argsort()[::-1]].reset_index(drop=True)
    # write dictionary
    n = list()
    for i in range(df_stats_signif.shape[0]):
        n.append(dict_chemID_to_metabolite[df_stats_signif.Metab[i]])
    df_stats_signif['Metabolite'] = n
    df_stats_signif.to_csv('../processed/stats_signif_{}_to_{}.csv'.format(group2,group1))
    
    # make volcano plots
    df_stats = df_stats.dropna(axis=0)
    idx_ns = df_stats.loc[df_stats.pvalue >= cutoff_val].index
    idx_sig = df_stats.loc[df_stats.pvalue < cutoff_val].index
    
    plt.figure()
    plt.title('{} to {}'.format(group2, group1))
    plt.xlabel('Log2 Fold Change')
    plt.ylabel('-log10(p-value)')
    plt.scatter(df_stats.loc[idx_ns,'LFC'], -np.log(df_stats.loc[idx_ns,'pvalue']),c='k')
    plt.scatter(df_stats.loc[idx_sig,'LFC'], -np.log(df_stats.loc[idx_sig,'pvalue']),c='r')
    plt.show()
    
```
