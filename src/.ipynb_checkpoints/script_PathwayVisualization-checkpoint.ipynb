# Find Fold-change, SNR of metabolites in pathway given temperature threshold

```python
temp_threshold=36
idx_sick = merged_data.loc[merged_data.Final_Temp<temp_threshold,:].index
idx_healthy = merged_data.loc[merged_data.Final_Temp>=temp_threshold,:].index
```

```python
# initialize results arrays
results_foldChange = np.empty(metabolite_nodes.shape[0])
results_SNR = np.empty(metabolite_nodes.shape[0])

for i in range(metabolite_nodes.shape[0]):
    # get the subset of the merged_data df that has the chemID of chosen metabolite
    chemID = metabolite_nodes.loc[i,'Chemical_ID']
    merged_data2 = merged_data.loc[:,'X'+str(chemID)]
    
    # get the values for sick, healthy using the indices generated before
    merged_data2_sick = merged_data2[idx_sick]
    merged_data2_healthy = merged_data2[idx_healthy]
    
    # compute fold-change, SNR
    FoldChange_val = np.mean(merged_data2_sick)/np.mean(merged_data2_healthy)
#     print(FoldChange_val)
    SNR_val = (np.mean(merged_data2_sick)-np.mean(merged_data2_healthy))/(np.std(merged_data2_sick)+np.std(merged_data2_healthy))
    
    # Store values
    results_foldChange[i]=FoldChange_val
    results_SNR[i]=SNR_val

# Append results to the metabolite_nodes dataframe
metabolite_nodes['FoldChange']=results_foldChange
metabolite_nodes['SNR']=results_SNR
```

# Construct Directed Network graph of pathway

```python
pathway_list = metabolite_nodes.Pathway_Name.unique()
```

```python
graph_array = []

for i in range(len(pathway_list)):
    pathway = pathway_list[i]
    print(pathway)
    pathway_nodes = metabolite_nodes.loc[metabolite_nodes['Pathway_Name']==pathway,:]
#     print(pathway_nodes.shape)
    pathway_edges = metabolite_edges.loc[metabolite_edges['Pathway_Name']==pathway,:]
#     print(pathway_edges)
    
    ########################################################
    # make network using pandas df edges
    G = nx.from_pandas_edgelist(pathway_edges,'Source','Destination',create_using=nx.DiGraph())
    
    # make dictionary of node index to metabolite name
    nodes_attr = pathway_nodes.set_index('Node_Idx').to_dict(orient='index')
    
    # set the  attribute to label nodes with the metabolite name
    nx.set_node_attributes(G,nodes_attr)

    # Store the graph
    graph_array.append(G)
#     print(G.nodes(data=True))
```

```python
# Visualize fold-changes
for i in range(len(pathway_list)):
    pathway = pathway_list[i]
    G = graph_array[i]
    # draw Graph
    labels = nx.get_node_attributes(G,'Metabolite')
    plt.figure(figsize=(10,10))
    plt.title('Log2(Fold-Change): '+pathway)
    node_colors = np.log2(list(nx.get_node_attributes(G,'FoldChange').values()))
#     print(list(nx.get_node_attributes(G,'FoldChange')))
#     print(node_colors)
    colormap = plt.cm.bwr
    nx.draw(G, labels=labels,font_size=8,pos=nx.circular_layout(G),
           node_color=node_colors,cmap=colormap,vmin=-2,vmax=2)
    ax=plt.gca()
    PCM=ax.get_children()[0] # this index may vary
    plt.colorbar(PCM, ax=ax)
    filename = '../processed/plot_network_LogFC_'+pathway+'.png'
    plt.savefig(filename,dpi=300, bbox_inches = "tight")
    plt.show()
```

```python
# Visualize SNR
for i in range(len(pathway_list)):
    pathway = pathway_list[i]
    G = graph_array[i]
    # draw Graph
    labels = nx.get_node_attributes(G,'Metabolite')
    plt.figure(figsize=(10,10))
    plt.title('SNR: '+pathway)
    node_colors = list(nx.get_node_attributes(G,'SNR').values())
#     print(list(nx.get_node_attributes(G,'FoldChange')))
#     print(node_colors)
    colormap = plt.cm.bwr
    nx.draw(G, labels=labels,font_size=8,pos=nx.circular_layout(G),
           node_color=node_colors,cmap=colormap,vmin=-2,vmax=2)
    ax=plt.gca()
    PCM=ax.get_children()[0] # this index may vary
    plt.colorbar(PCM, ax=ax)
    filename = '../processed/plot_network_SNR_'+pathway+'.png'
    plt.savefig(filename,dpi=300, bbox_inches = "tight")
    plt.show()
```
